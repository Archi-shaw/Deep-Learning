#Matrix multiplication
1. Element wise
2. Dot product

There are two main rules that performing matrix multiplication needs to satisfy:
1. The **inner dimensions** must match:
*`(3,2) @ (3,2)` won't work
* `(2,3) @ (3,2)` will work
*`(3,2) @ (2,3)` will work
2. The resulting matrix has the shape of the **outer dimensions**:
* `(2,3) @ (3,2)` -> `(2,2)`
*`(3,2) @ (2,3)` -> `(3,3)`



```python
torch.matmul(torch.rand(3,10), torch.rand(10,3))
```

## Reshaping, stacking, squezzing and unsequeezing tensors

* Reshaping - reshape an input to a defined shape
* View - Return a view of an input tensor of certain  shape but keep the same memory as the original tensor
* Stacking - Combine multiple tensors on top of each other (vstack) or side by side (htack)
* Squeezzing - removes all '1' dimensions from a tensor
* Unsequeezing - add a '1' dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted (Swapped) in a certain way

```python
# Let's create a tensor
import torch
x = torch.arange(1.,10.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1,9)
z,z.shape
```

```python
#Changing z changes x(because a view of a tensor ahares the same memory as the original input)
z[:,0] = 5
z,x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x,x,x,x], dim=0)
x_stacked
```

```python
# torch.sequeeze() - removes all single dimensional from a target tensor
x_reshaped
```

```python
x_reshaped.shape
```

```python
x_reshaped.squeeze()
```

```python
x_reshaped.squeeze().shape
```

```python
# torch.squeeze() - removes all 1D from a target tensor
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimension from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
# torch.unsqueeze() - adds a 1d to a target tensor at a specifix dim
print(f"Previous target: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

# Add an extra dimension with torch.unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=1)
print(f"\n New tensor: {x_unsqueezed}")
print(f"\n New shape: {x_unsqueezed.shape}")
```

```python
# torch.permute - rearranges the dimension of target tensor in a specified order
x_original = torch.rand(size=(224,224,3)) # [height,width,colur_channels]

# permute the original tensor to rearrange the axis order
x_permuted = x_original.permute(2,0,1)
print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
```

# Pytorch tensors & NUmpy

Numpy is a popular scientific Python numerical computing library.
And because of this, Pytorch has fuctionality to interact with it.

* Data in Numpy, want in Pytorch tensor -> torch.from_numpy(ndarray)

* Pytorch tensor -> Numpy -> torch.Tensor.numpy()

```python
# Numpy array to tensor

import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array).type(torch.float32)
array, tensor

# Making array from numpy will create data type of float 64 where converting that intotensor while
# will also result in data type of float64, we can change data taype to
# float 32 by using type(torch.float32) . If we make an array with tensor it
# will result in datatype of float 32. It will reflect numpy default data type of float 64
```

```python
array.dtype
```

```python
tensor.dtype
```

```python
# Change the value of array, what will this do to 'tensor'?
array = array+1
array, tensor
```

```python
# Tensor to Numpy array

tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

```python
# Change the tensor, what happens to 'numpy array'?
tensor = tensor + 1
array, numpy_tensor
```

# Reproducibility (trying to take random out of random )

In short how a neural network learns:


`start with random no -> tensors operation -> update random no to try and make them better reprsentations of the data -> again -> again...`


To reduce randomness in neural networks and Pytorch comes the concept of a  random seed

```python
torch.rand(3,3)
```

```python
import torch

# create a two random tensors
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# Let's make some random but reproducible tensors
import torch

# Set the random seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3,4)

print(random_tensor_C)
print(random_tensor_D)

print(random_tensor_C == random_tensor_D)
```
